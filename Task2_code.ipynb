```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import joblib

kf = RepeatedStratifiedKFold(n_splits=3, n_repeats=2, random_state=33)

accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for train_index, test_index in kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    random_forest = RandomForestClassifier(
        n_estimators=50,
        max_depth=10,
        min_samples_split=10,
        min_samples_leaf=5,
        max_features='sqrt',
    ).fit(X_train, y_train)

    y_pred = random_forest.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    accuracy_scores.append(accuracy)
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)

print("Metrics for Classification using Random Forest:")
print(f"Accuracy: {np.mean(accuracy_scores)}")
print(f"Precision: {np.mean(precision_scores)}")
print(f"Recall: {np.mean(recall_scores)}")
print(f"F1-score: {np.mean(f1_scores)}")
```

The data preprocessing and cleaning tasks focus on ensuring the accuracy and consistency of key variables in the loan dataset. The ActiveLateCategory and ActiveLateLastPaymentCategory are designed to categorize loans based on how overdue they are, using the principal debt and days since the last payment. This step ensures that overdue loans are flagged and categorized accurately. Additionally, the ActiveScheduleFirstPaymentReached field verifies whether the first payment has been made according to the active payment schedule. These tasks address potential inconsistencies in payment-related data.

Next, borrower-specific details are cleaned and standardized. For instance, the Age of borrowers is validated to ensure it accurately reflects their age at the time of loan application, while the Amount field ensures the loan balance recorded in the Primary and Secondary Markets is correct. Similarly, fields like AmountOfPreviousLoansBeforeLoan and AppliedAmount are verified for accuracy, confirming the correct amounts of previous loans and the initial applied amount. This step ensures the financial data tied to borrowers is correct and ready for analysis.

Auction-related data such as AuctionBidNumber, AuctionId, AuctionName, and AuctionNumber is also crucial. These fields must be cross-checked and cleaned to ensure that all auction records are uniquely identified and properly assigned to each loan. Additionally, the BidPrincipal field, which represents bid amounts in both Primary and Secondary Markets, is validated to ensure accurate recording. Other fields like BidsApi, BidsManual, BidsPortfolioManager must reflect the correct amounts of investments made through various methods, safeguarding the integrity of investment-related data.

Data labeling for the target variable is a critical step in preparing the dataset for model development. The task involves creating a binary target variable to identify loans that have defaulted and those that have not. If the DefaultDate field is populated, the entry is labeled as "Default"; otherwise, it is labeled as "Not Default." This labeling process is crucial for model training and evaluation, providing a clear indicator of loan performance outcomes.

Finally, additional fields such as location data and borrower financial information are cleaned for consistency. Fields like City, Country, and County are standardized, ensuring that location data is uniform. Credit Scores are categorized and cleaned according to predefined criteria, while fields like ContractEndDate, DebtToIncome, and EmploymentStatus are validated for accuracy. This ensures that all borrower-related data is properly prepared, improving the quality and reliability of the dataset for analysis and modeling.
